```python
target_terms = ("@BBC", "@CBS", "@FoxNews", "@nytimes", "@CNN")
```

```python
news_room = pd.DataFrame(columns = ['source_account', 'source_name', 'text', 'tweet_time', 'compound', 'positive', 'neutral', 'negative', 
                                    'tweet_datetime', 'time_in_between'])
```

```python
# Loop through all target users
for target in target_terms:
    
    tweet_times = []  
    #datetime = []
    # Run search around each tweet
    public_tweets = api.search(target, count=100, result_type="recent")

    # Loop through all tweets
    for tweet in public_tweets["statuses"]:

        # Run Vader Analysis on each tweet
        source = target
        text = tweet["text"]
        
        tweet_time = tweet["created_at"]
        #tweet_datetime = datetime.strptime(tweet["created_at"], "%a %b %d %H:%M:%S %z %Y")
        
        #datetime.append(tweet_datetime)
                
        tweet_times.append(tweet["created_at"])
        
        
        
        compound = analyzer.polarity_scores(tweet["text"])["compound"]
        pos = analyzer.polarity_scores(tweet["text"])["pos"]
        neu = analyzer.polarity_scores(tweet["text"])["neu"]
        neg = analyzer.polarity_scores(tweet["text"])["neg"]
        
         
        news_room = news_room.append({'source_account': source,
                                      'text':text,
                                      'tweet_time': tweet_time,
                                      'compound': compound, 
                                     'positive': pos,
                                     'neutral': neu,
                                     'negative': neg
                                     }, ignore_index=True)
        
            
        # Convert all tweet times into datetime objects
        tweet_time_objects = []

        # Add each datetime object into the array
        for x in range(len(tweet_times)):
            tweet_datetime = datetime.strptime(tweet_times[x], "%a %b %d %H:%M:%S %z %Y")
            tweet_time_objects.append(tweet_datetime)
            
        # Calculate the time between tweets
        time_in_between = []
        
        # Calculate the time in between each tweet
        
        for x in range(len(tweet_time_objects)-1):
            secs_apart = ((tweet_time_objects[x] - tweet_time_objects[x+1]).seconds)
            time_in_between.append(secs_apart)
            
        news_room = news_room.append({'tweet_datetime': tweet_datetime, 
                                      'time_in_between': time_in_between}, ignore_index=True)
        
        
        
news_room.head()
```

```python
news_room['tweet_datetime'] = pd.to_datetime(news_room['tweet_time'])
news_room['source_name']= news_room.replace({'source_account': {'@':''}}, regex=True)
news_room.head()
```
